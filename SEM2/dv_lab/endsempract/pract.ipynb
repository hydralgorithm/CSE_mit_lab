```python
import re

stop = set(["the","is","in","and","to","a","of"])
def preprocess(text):
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]','',text)
    tokens = text.split()
    return [t for t in tokens if t not in stop]

print(preprocess("This is an example!")) 
```
