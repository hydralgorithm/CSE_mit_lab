```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../datasets/northwind_order_details.csv")
df["Sales"] = df["quantity"] * df["unit_price"]
top = df.groupby("product_id")["Sales"].sum().nlargest(5)
print(top)
top.plot(kind="bar")
plt.show()
```
