```python
import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/cereals_dirty.csv")
print("Dataset Loaded!")

print(f"5-number summary b4 preprocessing:\n{df.describe().round(2)}\n")

numeric_cols = df.loc[ : ,"protein":"vitamins"].columns
for col in numeric_cols:
    valid_data = df[df[col] != -1][col]
    mean_val = valid_data.mean()
    df[col] = df[col].replace(-1,mean_val)

print(f"5-num summary after replacing missing values:\n{df[numeric_cols].describe().round(2)}\n")

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    median_val = df[col].median()
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - iqr*1.5
    upper = q3 + iqr*1.5
    df[col] = np.where((df[col]<lower) | (df[col]>upper), median_val, df[col])

print(f"5- num sum after treating noisy values:\n{df[numeric_cols].describe().round(2)}\n")
```
