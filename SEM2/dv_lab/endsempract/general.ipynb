```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/mtcars.csv")
print("DATASET LOADED!")

# BASIC INFO
print(f"Dataset Info:\n{df.info()}\n")
print(f"Dataset Statistical Summary:\n{df.describe().round(2)}\n")
# AVG Mileage
print(f"Average Mileage: {df["mpg"].mean():.2f}")
# Highest mileage car
print(f"Car with highest mileage: {df.loc[df["mpg"].idxmax(),"model"]}")
# Count of cars by cylinders
cyl_counts = df["cyl"].value_counts()
print(f"No. of cars by cylinders:\n{cyl_counts}\n")
# Visualizations
# Histo
plt.figure(figsize=(6,4))
plt.hist(df["mpg"], color="blue", bins=10, edgecolor="black")
plt.xlabel("Miles per gallon(mpg)")
plt.ylabel("Frequency")
plt.title("Distribution of mileage")
plt.grid(axis="y")
plt.show()
# Bar Chart
plt.figure(figsize=(6,4))
cyl_counts.plot(kind="bar",color="Olive",edgecolor="Black")
plt.xlabel("No. of cylinders")
plt.ylabel("Count")
plt.title("Cars by cylinder count")
plt.show()
# Scatter Plot
plt.figure(figsize=(6,4))
plt.scatter(df["wt"],df["mpg"])
plt.xlabel("Weight(1000lbs)")
plt.ylabel("Miles Per Gallon(mpg)")
plt.title("Weight vs Mileage")
plt.grid(True)
plt.show()
# Box Plot
plt.figure(figsize=(6,4))
plt.boxplot(df["mpg"],patch_artist=True)
plt.ylabel("Miles per Gallon")
plt.title("Boxplot of mpg")
plt.grid(axis="y")
plt.show()

print("\nAnalysis complete!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/cereals_nan.csv")
print("Dataset Loaded!")

numeric_cols = df.loc[:, "protein":"vitamins"].columns
print(f"5-number summary b4 preprocessing:\n{df[numeric_cols].describe().round(2)}\n")

for col in numeric_cols:
    mean_val = df[col].mean()
    df[col] = df[col].fillna(mean_val)
print(f"5-number summary after replacing missing values:\n{df[numeric_cols].describe().round(2)}\n")

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    median = df[col].median()
    upper = q3 + iqr * 1.5
    lower = q1 - iqr * 1.5
    df[col] = np.where((df[col] < lower) | (df[col] > upper), median, df[col])
print(f"5-number summary after cleaning noisy values:\n{df[numeric_cols].describe().round(2)}\n")

# Visualizations for numeric attributes
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col].dropna(), color="grey", edgecolor="black", bins=10)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.grid(axis="y")
    plt.show()

    plt.figure(figsize=(6, 4))
    plt.boxplot(df[col].dropna(), patch_artist=True)
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.grid(axis="y")
    plt.show()
```
