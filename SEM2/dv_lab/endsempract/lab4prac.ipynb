```python
import pandas as pd

temp = [32,12,39,27,23,21,25]
days = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
temp_series = pd.Series(temp,index=days)

print(f"Temp Series:\n{temp_series}\n")
# a)
print(f"Average temp of the week: {temp_series.mean():.2f}\n")
# b)
print(f"Maximum temperature of {temp_series.max()} at {temp_series.idxmax()}")
print(f"Minimum temperature of {temp_series.min()} at {temp_series.idxmin()}\n")
# c)
thresh = int(input("Enter threshold temperature: "))
thresh_temps = temp_series[temp_series>thresh]
print(f"Temp greater than threshold temp:\n{thresh_temps}\n")
# d)
fahren_temps = (9/5)*temp_series + 32;
print(f"Fahren Series:\n{fahren_temps}\n")
# e)
avg = temp_series.mean()
above_avgtemps = temp_series[temp_series>avg]
print(f"Above avg temp days:\n{list(above_avgtemps.index)}\n")
```

```python
import pandas as pd
data = {
    'Roll Number':[101,102,103,104,105,106,107,108,109,110],
    'Name':['Tim','Lee','Nemo','Lina','Jim','Bob','Mary','John','Alice','Shami'],
    'Gender':['M','M','M','F','M','M','F','M','F','F'],
    'Marks1': [78, 45, 66, 89, 34, 56, 90, 41, 72, 60],
    'Marks2': [65, 38, 70, 88, 42, 35, 91, 55, 39, 76],
    'Marks3': [80, 67, 72, 90, 45, 60, 85, 50, 78, 69]
    }

df = pd.DataFrame(data)
print(f"Student Dataframe:\n{df}\n")
# a)
df["Total Marks"] = df["Marks1"] + df["Marks2"] + df["Marks3"]
print(f"Df with total marks:\n{df}\n")
# b)
print(f"Lowest in Marks1: {df["Marks1"].min()}")
# c)
print(f"Highest in Marks2: {df["Marks2"].max()}")
# d)
print(f"Average in Marks3: {df["Marks3"].mean():.2f}")
# e)
avg_marks = df[["Marks1","Marks2","Marks3"]].mean(axis=1)
highest_stud = df.loc[avg_marks.idxmax(),"Name"]
print(f"\nStudent with highest average: {highest_stud}")
# f)
failed_students = (df["Marks2"]<40).sum()
print(f"\nNum of failed students: {failed_students}")
```
