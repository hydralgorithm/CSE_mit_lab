```python
import pandas as pd
import numpy as np

df = pd.read_csv("prac_datasets/cereals.csv")
print("Dataframe read successfully!")

df.describe()

numeric_cols = df.loc[:, "protein":"vitamins"].columns

print(f"5-number summary before pre-processing: ")
print(df[numeric_cols].describe().round(2))

for col in numeric_cols:
    valid_data = df[df[col] != -1][col]
    mean_val = valid_data.mean()
    df[col] = df[col].replace(-1, mean_val)

print(f"5-num summary after replacing missing values: \n{df.describe().round(2)}")

for col in numeric_cols:
    median_val = df[col].median()
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - iqr*1.5
    upper = q3 + iqr*1.5
    df[col] = np.where((df[col] < lower) | (df[col] > upper), median_val, df[col])

print(f"5-num summary after treating noisy values: \n{df.describe().round(2)}")
```
