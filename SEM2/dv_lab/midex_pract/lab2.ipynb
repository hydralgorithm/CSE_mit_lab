```python
numlist = []
n = int(input("Enter number of numbers you want to input in list: "))

for i in range(n):
    numlist.append(int(input(f"Enter num {i+1} : ")))
    
print(f"Displaying list:\n{numlist}")

maximum = max(numlist)
minimum = min(numlist)
total = 0
for num in numlist:
    total += num

avg = total/len(numlist)

print(f"\nMaximum number: {maximum}\nMinimum Number: {minimum}\nSum: {total}\nAverage: {avg:.2f}")

occ = int(input("Enter number to find occurences of: "))

# occ_count = 0;
# for num in numlist:
#     if occ == num:
#         occ_count +=1

occ_count = numlist.count(occ)
print(f"Occurences of {occ} = {occ_count}")
```

```python
sentence = input("Enter sentence: ").lower()
freq_counts = {}

words = sentence.split()
for word in words:
    freq_counts[word] = freq_counts.get(word, 0) + 1

print(f"Word frequencies: {freq_counts}")
```

```python
set1 = set()
set2 = set()

n1 = int(input("Enter number of elements in set 1: "))
print("Entering numbers for set 1: ")
for i in range(n1):
    set1.add(int(input(f"Enter element {i+1}: ")))

n2 = int(input("Enter number of elements in set 2: "))
print("Entering numbers for set 2: ")
for i in range(n2):
    set2.add(int(input(f"Enter element {i+1}: ")))

print(f"Set1 : {set1}")
print(f"Set2 : {set2}")

union = set1 | set2
intersection = set1 & set2
diff1 = set1 - set2
diff2 = set2 - set1

print(union)
print(intersection)
print(diff1)
print(diff2)
```
