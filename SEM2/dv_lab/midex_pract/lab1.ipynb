```python
def calc():
    num1 = float(input("Enter num1: "))
    num2 = float(input("Enter num2: "))
    print("Sum = ",(num1+num2))
    print("Difference = ",(num1-num2))
    print("Product = ",num1*num2)
    if(num2 != 0):
        print(f"Quotient = {(num1/num2):.2f}")
    else:
        print("Not divisible by 0!")

calc()
```

```python
def isPrime(num):
    if num < 2:
        return False
    for i in range(2,num//2 + 1):
        if num%i == 0:
            return False
    return True


    
n = int(input("Enter number: "))
result = isPrime(n)
if result:
    print("Prime")
else:
    print("Not prime")
```

```python
def sequenceGen():
    evensum = 0
    for i in range(1,101):
        if i%2 != 0:
            print(i, end=" ")
        else:
            evensum += i
    print()
    print(f"Sum of even numbers: {evensum}")

sequenceGen()
```
