```python
import pandas as pd
import matplotlib.pyplot as plt

od = pd.read_csv('../datasets/northwind_order_details.csv')
od['Sales'] = od['quantity'] * od['unit_price']
top = od.groupby('product_id')['Sales'].sum().nlargest(5)
top.plot(kind='bar')
plt.ylabel('Total Sales')
plt.title('Top 5 Products by Total Sales')
plt.savefig('../datasets/nw_top5_products.png')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

orders = pd.read_csv('../datasets/northwind_orders.csv', parse_dates=['order_date'])
od = pd.read_csv('../datasets/northwind_order_details.csv')
od['Sales'] = od['quantity'] * od['unit_price']
merged = orders.merge(od, on='order_id')
monthly = merged.set_index('order_date').resample('MS')['Sales'].sum()
monthly.plot()
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.savefig('../datasets/nw_monthly_sales.png')
plt.show()
```
