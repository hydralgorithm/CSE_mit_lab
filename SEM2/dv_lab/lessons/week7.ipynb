```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/ipl_batting.csv')
top = df.groupby('Player')['Runs'].sum().nlargest(5)
top.plot(kind='bar')
plt.ylabel('Total Runs')
plt.title('Top 5 run scorers')
plt.savefig('../datasets/ipl_top5_runs.png')
```

```python
import pandas as pd
df_matches = pd.read_csv('../datasets/ipl_matches.csv') # ensure 'winner' column exists
winners = df_matches['winner'].value_counts()
print('Team with most wins:', winners.idxmax(), 'Count:', winners.max())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('../datasets/ipl_matches.csv') # ensure 'bat_first' and 'total_runs' columns or similar
# Example assumes a column 'batting_order' with values 'first'/'second'
sns.boxplot(x='batting_order', y='total_runs', data=df)
plt.title('Scores by batting order')
plt.savefig('../datasets/ipl_scores_boxplot.png')
```

```python
import pandas as pd

df = pd.read_csv('../datasets/ipl_batting.csv')
top10 = df.nlargest(10, 'StrikeRate')[['Player', 'StrikeRate']]
print(top10.to_string(index=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../datasets/ipl_matches.csv')
extras_by_team = df.groupby('team1')['extras'].sum().sort_values(ascending=False)
extras_by_team.plot(kind='bar')
plt.title('Extras conceded by team')
plt.savefig('../datasets/ipl_extras_by_team.png')
```
