```python
numList = []
n = int(input("Enter number of numbers you want to add to list: "))

for i in range(0, n):
    numList.append(int(input("Enter a number to add to list: ")))

total = 0
lar = numList[0]
small = numList[0]
for num in numList:
    total += num
    if num >= lar:
        lar = num
    if num < small:
        small = num

avg = total/n

print(f"Largest number = {lar}")
print(f"Smallest number = {small}")
print(f"Sum of numbers = {total}")
print(f"Average = {avg}")

num_occur = int(input("Enter number you want to find occurences of: "))
occurences = numList.count(num_occur)
print(f"The number {num_occur} occurs {occurences} times.")
```

```python
winners = {
    2003: "Australia",
    2007: "Australia",
    2011: "India",
    2015: "Australia",
    2019: "England",
    2023: "India"
}

unique_countries = sorted(set(winners.values()))
print("Unique list of World Cup winners:", unique_countries)

counts = {}
for country in winners.values():
    if country in counts:
        counts[country] += 1
    else:
        counts[country] = 1

max_wins = 0
for c in counts.values():
    if c > max_wins:
        max_wins = c

best_countries = []
for country, cnt in counts.items():
    if cnt == max_wins:
        best_countries.append(country)

if len(best_countries) == 1:
    print(f"Best-performing country: {best_countries[0]} ({max_wins} wins)")
else:
    print(f"Best-performing countries: {', '.join(best_countries)} ({max_wins} wins each)")
```
