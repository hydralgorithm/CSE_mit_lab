```python
num1 = float(input("Enter num1: "))
num2 = float(input("Enter num2: "))
sum = num1 + num2
difference = num1 - num2
product = num1 * num2
print(f"Sum = {sum:.2f}")
print(f"Difference = {difference:.2f}")
print(f"Product = {product:.2f}")
if num2 != 0:
    div = num1 / num2
    print(f"Division = {div:.2f}")
else:
    print("Not divisible!")
```

```python
def isPrime(num): 
    if num < 2:
        return False   
    for i in range(2,num//2+1):
        if num % i == 0:
            return False     
    return True
    
num = int(input("Enter number: "))
print(isPrime(num))
```

```python
sum_even = 0
for i in range(1,101):
    if i%2 != 0:
        print(i, end=" ")
    else:
        sum_even += i

print()
print(f"Sum of even numbers: {sum_even}")
```

```python
def addElements(a,b):
    result = a + b
    print(f"Result: {result}")

addElements(5,3)
addElements(6.7,4.1)
addElements("Hey ","Guys!")
```

```python
inputVal = input("Enter string: ").lower()
cons = 0
vow = 0

for char in inputVal:
    if char in 'aeiou':
        vow += 1
    elif char.isalpha():
        cons += 1

print(f"Number of Vowels: {vow}")
print(f"Number of Consonants: {cons}")
```
